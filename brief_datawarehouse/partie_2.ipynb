```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('mysql+pymysql://root:password@127.0.0.1:3306/dwh_sales')
conn = engine.connect()
```

```python
query = '''
    SELECT 
        s.SupplierID,
        s.SupplierLocation,
        
        sh.ShipperID,
        sh.ShippingMethod,

        p.ProductID,
        p.ProductName,
        p.ProductCategory,
        p.ProductSubCategory,
        p.ProductPrice,

        c.CustomerID,
        c.CustomerSegment,

        d.DateID,
        d.Annee,
        d.Mois,
        d.Jour,

        f.id,
        f.QuantitySold,
        f.TotalAmount,
        f.DiscountAmount,
        f.NetAmount,

        fs.id,
        fs.StockReceived,
        fs.StockSold,
        fs.StockOnHand

    FROM
        FactVentes f
    JOIN
        Supplier s ON f.SupplierID = s.SupplierID
    JOIN
        Shipping sh ON f.ShipperID = sh.ShipperID
    JOIN
        Product p ON f.ProductID = p.ProductID
    JOIN
        Customer c ON f.CustomerID = c.CustomerID
    JOIN
        Date d ON f.DateID = d.DateID
    LEFT JOIN
        FactStock fs ON p.ProductID = fs.ProductID AND f.DateID = fs.DateID
    ;
'''

df = pd.read_sql_query(query, conn)
df.dropna()
```

```python
df.to_csv('data/data_export.csv')
```
